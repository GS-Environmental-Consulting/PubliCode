```python
import pandas as pd
import numpy as np
```

### Paramaters

```python
# Set up inputs
capex_period = 12  # Monthly
opex_period = 4    # Quarterly
project_length = 30 #years
starting_date= '2022-01-01'
income_types = ['Reputation', 'Advertisement', 'Water_retention', 'Flood_prev', 'Type 5',
                'Type 6', 'Type 7', 'Type 8', 'Type 9', 'Type 10']
revenues = pd.Series(np.random.randint(100000, 1000000, size=project_length*12))
tax_rate = 0.3
depreciation_rate = 0.1
```

### Dataframe

```python
# Set up dataframe
df = pd.DataFrame(columns=['Period', 'Revenue'] + income_types + ['Opex', 'Depreciation', 'EBIT', 'Taxes', 'Net Income'])
df['Period'] = pd.date_range(start=starting_date, periods=project_length*12, freq='Q')
df['Revenue'] = revenues
df[income_types] = pd.DataFrame(np.random.randint(0, 10000, size=(project_length*12, len(income_types))), columns=income_types)
```

```python
display (df)
```

### Calculations

```python
# Calculate Opex
#opex = pd.Series(np.random.randint(5000, 50000, size=capex_period*opex_period))
#df['Opex'] = np.repeat(opex, 3)[:capex_period*12]

# Calculate Depreciation
df['Depreciation'] = df['Revenue'] * depreciation_rate

# Calculate EBIT
df['EBIT'] = df['Revenue'] - df[income_types].sum(axis=1) - df['Opex'] - df['Depreciation']

# Calculate Taxes
df['Taxes'] = df['EBIT'] * tax_rate

# Calculate Net Income
df['Net Income'] = df['EBIT'] - df['Taxes']
```
